To create a confidence interval we need to calculate the standard error of the mean (SEM). By using the formula $SEM = \frac{σ}{\sqrt(n)}$. After calculating the standard error of the mean we can multiply the SEM by 1.96 because under a normal distribution 95% of the values are within 1.96 standard deviation of the mean. This gives us the margin of error, and after we finally can calculate the Confidence interval by multiplying the mean by the margin of error. The confidence interval of 95% means that if we were to take 1000 different random samples and compute them into 1000 different confidence intervals, we would see 950 of those intervals to contain the true population parameter.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Create a sample dataset of 100 data points from a normal distribution
sample_data = np.random.normal(loc=100, scale=15, size=100)  # Mean=100, SD=15

def bootstrap_ci(data, num_bootstraps=1000, confidence_level=0.95, statistic=np.mean):
    """
    Calculate the bootstrap confidence interval for a given statistic.

    Parameters:
    - data: The original sample data.
    - num_bootstraps: The number of bootstrap samples to generate.
    - confidence_level: The desired confidence level for the interval.
    - statistic: The statistic to calculate (mean, median, etc.).

    Returns:
    - ci: A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Generate bootstrap samples and compute the specified statistic for each sample
    bootstrapped_stats = np.array([statistic(np.random.choice(data, size=len(data), replace=True)) for _ in range(num_bootstraps)])
    
    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrapped_stats, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrapped_stats, (1 + confidence_level) / 2 * 100)
    
    return lower_bound, upper_bound

# Calculate the 95% bootstrap confidence interval for the population mean
mean_ci = bootstrap_ci(sample_data, num_bootstraps=1000, confidence_level=0.95, statistic=np.mean)

# Calculate the 95% bootstrap confidence interval for the population median
median_ci = bootstrap_ci(sample_data, num_bootstraps=1000, confidence_level=0.95, statistic=np.median)

# Output the confidence intervals
mean_ci, median_ci
````

The bootstrapping consisted of using the sample's health scores to compute multiple different means over multiple different simulations. Now we can see the graph of these said means.

````python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Calculate the difference between FinalHealthScore and InitialHealthScore
vaccine_df['HealthScoreChange'] = vaccine_df['FinalHealthScore'] - vaccine_df['InitialHealthScore']

# Define the bootstrap function
def bootstrap_means(data, n_bootstrap=1000, seed=None):
    if seed is not None:
        np.random.seed(130)  # Set the seed for reproducibility
    boot_means = []
    for _ in range(n_bootstrap):
        # Resample with replacement
        sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the sample
        boot_means.append(np.mean(sample))
    return boot_means

# Perform bootstrapping on HealthScoreChange with a seed of 130
bootstrapped_means = bootstrap_means(vaccine_df['HealthScoreChange'].values, seed=130)

# Convert the bootstrapped means into a DataFrame for plotting
bootstrapped_df = pd.DataFrame(bootstrapped_means, columns=['BootstrappedMean'])

# Calculate the standard error of the mean (SEM)
sem = np.std(bootstrapped_means) / np.sqrt(len(bootstrapped_means))

# Calculate the 95% confidence interval
ci_lower = np.percentile(bootstrapped_means, 2.5)
ci_upper = np.percentile(bootstrapped_means, 97.5)

# Print the SEM and confidence intervals
print(f"Standard Error of the Mean (SEM): {sem:.4f}")
print(f"95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")

# Create a Plotly histogram of bootstrapped means
fig = px.histogram(bootstrapped_df, x='BootstrappedMean',
                   labels={'BootstrappedMean': 'Bootstrapped Mean of Health Score Changes'},
                   title='Distribution of Bootstrapped Means with 95% Confidence Interval',
                   nbins=30,  # Adjust the number of bins for smoothness
                   color_discrete_sequence=['orange'])

# Add the confidence interval lines
fig.add_shape(type="line",
              x0=ci_lower, y0=0, x1=ci_lower, y1=10,  # Adjust y1 for visibility
              line=dict(color="red", width=2, dash="dash"),
              name="95% CI Lower Bound")

fig.add_shape(type="line",
              x0=ci_upper, y0=0, x1=ci_upper, y1=10,  # Adjust y1 for visibility
              line=dict(color="blue", width=2, dash="dash"),
              name="95% CI Upper Bound")

# Add annotations for confidence intervals
fig.add_annotation(x=ci_lower, y=0, text=f"95% CI Lower: {ci_lower:.2f}", 
                   showarrow=True, arrowhead=2, ax=0, ay=-40, font=dict(color="red"))
fig.add_annotation(x=ci_upper, y=0, text=f"95% CI Upper: {ci_upper:.2f}", 
                   showarrow=True, arrowhead=2, ax=0, ay=-40, font=dict(color="blue"))

# Add annotation for SEM
sem_x_position = np.mean(bootstrapped_means)  # Position SEM annotation at mean of bootstrapped means
fig.add_shape(type="line",
              x0=sem_x_position, y0=0, x1=sem_x_position, y1=10,  # Adjust y1 for visibility
              line=dict(color="green", width=2, dash="dash"),
              name="Standard Error of the Mean")

fig.add_annotation(x=sem_x_position, y=0, text=f"SEM: {sem:.4f}", 
                   showarrow=True, arrowhead=2, ax=0, ay=-40, font=dict(color="green"))

# Show the plot
fig.show(renderer='png')


````

**Proposal 1:**
- Do students perform worse academically when they play video games frequently?

Variables:
- Video Game Time (Continuous Variable): How many hours spent playing video games.
  - Visualization: Scatter plot to show the relationship between playing video games and exam results. This would show any trends or correlations.
- Exam Scores (Continuous variable): The scores received on the exam.
  - Visualization: Histogram to display the distribution of all exam scores, showing overall performance of all students.

Analysis:
- A linear regression analysis will be used to see the relationship between playing video games and exam scores. The assumptions for linear regression include normal distribution, homoscedasticity, linearity, and independence.

Hypothesis:
- Null Hypothesis: There is no significant relationship between playing video games and exam scores
- Alternative Hypothesis: There is a significant negative relationship between playing video games and exam scores
- If the results show that the p-value < 0.05 then it would suggest that playing video games negatively affects academic performance.

**Proposal 2:**

- Hours Worked (Continuous Variable): Total number of hours worked on assignments, projects, and studying.
  - Visualization: Box plot to display the distribution of hours worked for both groups. Showing difference in workloads.

Analysis: 
- We will use linear regression where hours worked per week is the dependent variable, and the student group (STEM or Business) is the independent variable. The assumptions for linear regression include normal distribution, homoscedasticity, linearity, and independence.

Hypothesis:
- Null Hypothesis: There is a significant difference in workload between STEM students and Business students.
- Alternative Hypothesis: STEM students have a greater workload than business students.
- If the results show that the p-value < 0.05 then it would suggest that STEM students have a greater workload.
